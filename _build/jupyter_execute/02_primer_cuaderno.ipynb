```python
# prompt: Hola, cargame los datos del archivo ifood_df.csv para hacer un análisis de datos

import pandas as pd

# Load the dataset from the user's Google Drive
df = pd.read_csv("ifood_df.csv")

# Show DataFrame
df.head(5)
```

```python
# prompt: Ayudame a ver las filas de la 20 a la 30 de df

df.iloc[20:31]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asegúrate de que 'Income' es un número
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')

# Descripción de la variable 'Income'
print(df['Income'].describe())

# Histograma de la variable 'Income'
plt.figure(figsize=(10, 6))
sns.histplot(df['Income'], kde=True)
plt.title('Distribución de Income')
plt.show()

# Boxplot de la variable 'Income'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Income'])
plt.title('Boxplot de Income')
plt.show()

```

```python
# prompt: Analiza la variable Income: describela y crea un histograma

# Describe the Income variable
print(df['Income'].describe())

# Create a histogram of the Income variable
import matplotlib.pyplot as plt
plt.hist(df['Income'])
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Histogram of Income')
plt.show()
```
